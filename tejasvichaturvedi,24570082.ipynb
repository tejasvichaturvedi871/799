```python
#A create a simple tensor
import tensorflow as tf

x = tf.constant([[10, 20], [30, 40]])
x
```

```python
#Basic tensor operation
a = tf.constant([5, 10])
b = tf.constant([2, 4])

print("Add:", tf.add(a, b))
print("Multiply:", tf.multiply(a, b))
print("Subtract:", tf.subtract(a, b))
print("Divide:", tf.divide(a, b))
```

```python
#Variable Tensor
v = tf.Variable([1, 2, 3])
print("Before:", v.numpy())

v.assign([4, 5, 6])
print("After:", v.numpy())
```

```python
#numpy to tensor
import numpy as np

arr = np.array([[1,2,3]])
tensor = tf.convert_to_tensor(arr)
tensor
```

```python
#Dense layer output
layer = tf.keras.layers.Dense(3, activation="relu")
output = layer(tf.constant([[1., 2., 3.]]))
output
```

```python
#Sequential model summary
model = tf.keras.Sequential([
    tf.keras.layers.Dense(5, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.summary()
```

```python
#model compile
model.compile(optimizer="adam", loss="mse", metrics=["mae"])
print("Model Compiled Successfully!")
```

```python
#Train on random data
import numpy as np

X = np.random.rand(100, 3)
y = np.random.rand(100, 1)

model.fit(X, y, epochs=5)
```

```python
#Predit
test_data = np.array([[0.3, 0.6, 0.9]])
model.predict(test_data)
```
